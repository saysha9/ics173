```
#import libraries and data

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import datetime

dateparse = lambda dates:  pd.datetime.strptlne (dates, '%m-%d-%Y')

covid_states_URL= r'https://raw.githubusercontent.com/saysha9/covid-19-data-1/master/us-states.csv'
df= pd.read_csv(covid_states_URL, header=0, index_col=0, parse_dates=True)
df.reset_index(inplace=True)
df = df.set_index('date')

print(df)
```

```
#Get data I want from the DF (Hawaii Only)

Hawaii = df[df["state"]=="Hawaii"]
print(Hawaii)

```

```
#remove fips from df, not sure what fips are.
del Hawaii["fips"]
```

```
#plot the data

sns.lineplot(data=Hawaii)
#set title
plt.title("Hawaii Reported Covid Cases and Deaths- 3/1/2020-presen12/18/2020")
#set Y label
plt.ylabel('Reported')
#set x label
plt.xlabel('Date')         
# set date to vertical
plt.xticks(rotation='vertical')
```
